```python
%sql
USE CATALOG job_catalog;
USE SCHEMA default;
```

```python
%sql
create or replace table silver_customers_orders as
select
  c.customer_id,
  c.customer_name,
  c.address,
  o.order_number,
  o.order_datetime,
  o.number_of_line_items
from bronze_customers c
join bronze_orders o
on c.customer_id = o.customer_id;
```

```python
%sql
-- Adding the 'state' column as I will need it later for the For Each iterator
ALTER TABLE silver_customers_orders
ADD COLUMN state STRING;

UPDATE silver_customers_orders
SET state = RIGHT(address, 2);
```
