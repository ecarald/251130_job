```python
%sql
USE CATALOG job_catalog;
USE SCHEMA default;

CREATE OR REPLACE TABLE silver_customers_sales AS
SELECT
  c.customer_id,
  c.customer_name,
  s.product_name,
  s.order_date, 
  s.product_category,
  s.total_price
FROM bronze_customers c
JOIN bronze_sales s
ON c.customer_id = s.customer_id;
```

```python
%sql
select * from silver_customers_sales;
```

```python
%sql
select distinct * from silver_customers_sales;
```

##Check for duplicates
This will be later required for the If/Else block

```python
df = spark.sql("select * from default.silver_customers_sales")

# Check for duplicates records
#OBS! Interesante forma de comprobar si existen duplicados
duplicates_exists = df.count() > df.dropDuplicates().count() #boolean value outcome

# Set boolean flag in taks values
dbutils.jobs.taskValues.set(key="has_duplicates", value=duplicates_exists)
```
